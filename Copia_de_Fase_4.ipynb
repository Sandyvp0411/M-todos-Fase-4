PASO 1: Revision Bibliografica - Fundamentacion Teorica

Bibliometria: Es el conjunto de metodos cuantitativos enfocados al analisis de la produccion cientifica a fin de evaluar tendencias, estructuras de conocimeito y dinamicas de investigacion espeficica. De acuerdo con Aria y Cuccurullo (2017), esta facilita el estudio de grandes volumenes de informacion con indicadores como autores, instituciones, revistas y tematicas de mayor impacto. Como herramienta, Muñoz (2026) destaca Bibliometrix por su capacidad para identificar aquellas tendencias emergentes, areas de investigacion consolidadas y oportunidades de innovacion.

Natural Language Processing (NLP): Disciplina que combina la linguistica computacional, inteligencia artificial y el aprendizaje automatico para comprender, interpretar y generar lenguaje humano. Young et al. (2018) indica que este ha evolucionado gracias a los avances Deep Learning, asi mismo, Surpriyono et al. (2024) destaca que sus tecnicas mejoran en tareas de clasificacion de textos, traduccion automatica, analisis de sentimientos, reconocimiento de entidades y generacion automatica de lenguaje.

Funciones de Densidad de Probabilidad (PDF): es una funcion matematica, representada comunmente como F(x), que describe la probabilidad relativa de que una variable aleatoria continua tome un valor predeterminado.

Machine Learning (ML): Es el conjunto de tecnicas y algoritmos que permiten a las computadoras aprender patrones directamente a partir de los datos, en lugar de ser programadas explicitamente con reglas logicas manuales.

Large Language Mode (LLM): Naveed et al. (2023), los define como modelos entrenados sobre grandes volumenes de datos textuales capaces de comprender, generar y analizar lenguaje natural con niveles de desempeño cercanos a los humanos.

```python
# =========================================================
# PASO 5: Análisis Bibliométrico en Python
# =========================================================

# 1. Cargar tus datos (Asegúrate de cambiar el nombre al de tu archivo CSV de Scopus)
df_scopus = pd.read_csv('Base Scopus.csv')

# Mostrar las columnas disponibles para ubicar los nombres correctos
print("Columnas disponibles en tu archivo:")
print(df_scopus.columns.tolist())
```

```python
# =========================================================
# PASO 5: Gráficos de Análisis Bibliométrico
# =========================================================

# Configurar estilo de los gráficos
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 5))

# --- GRÁFICO 1: Evolución de Publicaciones por Año ---
plt.subplot(1, 2, 1)
publicaciones_por_ano = df_scopus['Year'].value_counts().sort_index()
sns.lineplot(x=publicaciones_por_ano.index, y=publicaciones_por_ano.values, marker="o", color="royalblue", linewidth=2.5)
plt.title('Evolución Anual de Publicaciones sobre IA y Productividad', fontsize=12, fontweight='bold')
plt.xlabel('Año', fontsize=10)
plt.ylabel('Cantidad de Artículos', fontsize=10)

# --- GRÁFICO 2: Top 10 Palabras Clave más Frecuentes ---
plt.subplot(1, 2, 2)

# Limpiamos y separamos las palabras clave (vienen separadas por punto y coma ';')
keywords_series = df_scopus['Author Keywords'].dropna().str.split(';')
all_keywords = [kw.strip().lower() for sublist in keywords_series for kw in sublist if kw.strip() != ""]

# Contamos las frecuencias
top_keywords = pd.Series(all_keywords).value_counts().head(10)

# Graficamos las palabras clave
sns.barplot(x=top_keywords.values, y=top_keywords.index, hue=top_keywords.index, palette="viridis", legend=False)
plt.title('Top 10 Palabras Clave de los Autores', fontsize=12, fontweight='bold')
plt.xlabel('Frecuencia', fontsize=10)
plt.ylabel('Palabra Clave', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# =========================================================
# PASO 6: Análisis de Sentimiento (NLP) en los Resúmenes
# =========================================================
from nltk.sentiment import SentimentIntensityAnalyzer

# Inicializar el analizador de sentimientos de VADER
sia = SentimentIntensityAnalyzer()

# Asegurar que no haya valores nulos en la columna Abstract reemplazándolos por texto vacío
df_scopus['Abstract_Clean'] = df_scopus['Abstract'].fillna("")

# Definir función para calcular el puntaje de sentimiento compuesto (-1 a 1)
def calcular_sentimiento(texto):
    if texto == "":
        return 0.0
    return sia.polarity_scores(texto)['compound']

# Aplicar la función a cada abstract del DataFrame
df_scopus['Sentiment_Score'] = df_scopus['Abstract_Clean'].apply(calcular_sentimiento)

# Clasificar el sentimiento cuantitativo en categorías cualitativas
def clasificar_sentimiento(score):
    if score >= 0.05:
        return 'Positivo'
    elif score <= -0.05:
        return 'Negativo'
    else:
        return 'Neutro'

df_scopus['Sentiment_Category'] = df_scopus['Sentiment_Score'].apply(clasificar_sentimiento)

# --- Graficar la distribución de sentimientos ---
plt.figure(figsize=(6, 4))
sns.countplot(data=df_scopus, x='Sentiment_Category', hue='Sentiment_Category', palette={'Positivo': 'mediumseagreen', 'Neutro': 'orange', 'Negativo': 'crimson'}, legend=False)
plt.title('Distribución de Sentimientos en la Academia sobre IA', fontsize=12, fontweight='bold')
plt.xlabel('Categoría de Sentimiento', fontsize=10)
plt.ylabel('Cantidad de Artículos', fontsize=10)
plt.show()

# Mostrar un resumen rápido de los resultados en texto
print(df_scopus['Sentiment_Category'].value_counts(normalize=True) * 100)
```

```python
# =========================================================
# PASO 7: Predicción con Machine Learning (Series de Tiempo)
# =========================================================
from sklearn.linear_model import LinearRegression
import scipy.stats as stats

# 1. Preparar los datos históricos de publicaciones por año
df_anos = df_scopus['Year'].value_counts().reset_index()
df_anos.columns = ['Ano', 'Publicaciones']
df_anos = df_anos.sort_values(by='Ano')

# Filtrar años recientes si hay datos atípicos o incompletos (ej. años muy antiguos con 0 publicaciones)
df_anos = df_anos[df_anos['Ano'] >= 2015]

# 2. Preparar variables para el modelo de Machine Learning
X = df_anos['Ano'].values.reshape(-1, 1) # Variable independiente (Año)
y = df_anos['Publicaciones'].values      # Variable dependiente (Publicaciones)

# Entrenar el modelo de Regresión Lineal
modelo = LinearRegression()
modelo.fit(X, y)

# 3. Predecir los próximos 5 años (ej. de 2026 a 2030)
anos_futuros = np.array(range(2026, 2031)).reshape(-1, 1)
predicciones = modelo.predict(anos_futuros)

# Redondear predicciones ya que no se pueden publicar "fracciones" de artículos
predicciones_enteras = np.round(predicciones).astype(int)

# --- Visualización del Modelo Predictivo ---
plt.figure(figsize=(12, 5))

# Gráfico de la tendencia y predicción
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='darkblue', label='Datos Históricos (Reales)', s=60)
plt.plot(X, modelo.predict(X), color='royalblue', linestyle='--', label='Línea de Tendencia')
plt.scatter(anos_futuros, predicciones_enteras, color='crimson', label='Predicción ML (Futura)', marker='^', s=80)
plt.title('Predicción de Publicaciones Futuras (ML)', fontsize=12, fontweight='bold')
plt.xlabel('Año')
plt.ylabel('Cantidad de Artículos')
plt.legend()

# --- Distribución Estadística: Función de Densidad de Probabilidad (PDF) ---
plt.subplot(1, 2, 2)
# Ajustamos los datos históricos a una distribución normal para calcular la PDF
mu, std = stats.norm.fit(y)
x_pdf = np.linspace(min(y) - 10, max(y) + 20, 100)
p_pdf = stats.norm.pdf(x_pdf, mu, std)

plt.plot(x_pdf, p_pdf, 'k', linewidth=2, color='darkgreen')
plt.hist(y, bins=5, density=True, alpha=0.6, color='mediumseagreen')
plt.title('Densidad de Probabilidad (PDF) de Publicaciones', fontsize=12, fontweight='bold')
plt.xlabel('Volumen de Publicaciones Anuales')
plt.ylabel('Densidad')

plt.tight_layout()
plt.show()

# Imprimir las predicciones en texto para usarlas en el reporte
print("--- PREDICCIONES DEL MODELO DE MACHINE LEARNING ---")
for ano, pred in zip(anos_futuros.flatten(), predicciones_enteras):
    print(f"Año {ano}: Se proyectan aproximadamente {pred} publicaciones.")
```

A partir del análisis de datos consolidado en nuestro Dashboard, diseñamos las siguientes estrategias corporativas para guiar a las organizaciones en la adopción segura de la Inteligencia Artificial:

### 📈 Estrategia 1: Programa de Adaptación y Alfabetización en IA (Respuesta al NLP)
*   **Justificación:** El análisis de sentimiento detectó que un porcentaje importante de investigaciones advierte sobre la resistencia al cambio y barreras humanas en la implementación tecnológica.
*   **Acción:** Diseñar talleres de capacitación práctica y programas de "upskilling" enfocados en el uso seguro de herramientas de IA generativa para los empleados administrativos y operativos.
*   **KPI Asociado:**
    $$\text{Tasa de Adopción de Herramientas} = \left( \frac{\text{Colaboradores activos usando herramientas de IA}}{\text{Total de la plantilla capacitada}} \right) \times 100$$
*   **Meta:** Alcanzar el 75% de uso activo de herramientas licenciadas de IA en los primeros 6 meses post-capacitación.

### ⚙️ Estrategia 2: Automatización de Procesos de Negocio basada en Tendencias Científicas (Respuesta a Bibliometría)
*   **Justificación:** Las palabras clave dominantes demuestran que las sub-tecnologías con mayor madurez y validación científica actual son el "Machine Learning" y la automatización inteligente.
*   **Acción:** Priorizar la inversión en la automatización de los procesos de servicio al cliente (chatbots inteligentes de última generación) y análisis de inventario predictivo.
*   **KPI Asociado:**
    $$\text{Eficiencia Operativa en Procesos (EOP)} = \left( \frac{\text{Tiempo promedio de proceso con IA}}{\text{Tiempo promedio de proceso manual anterior}} \right) \times 100$$
*   **Meta:** Reducir el tiempo de atención y procesamiento de solicitudes de clientes en un 40% durante el primer año.
